```python
import pandas as pd
import csv
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('league_data.csv')
data.tail()
```

```python
scaler = StandardScaler()
scaler.fit(data)
scaled = scaler.fit_transform(data)

# Normalizing all Values
scaled_data = pd.DataFrame(scaled,columns=data.columns)
scaled_data.drop('blueWins', inplace=True, axis=1)

# Adding Back blueWins 
scaled_data.insert(0,'blueWins', data['blueWins'])
scaled_data['blueWins'] = scaled_data['blueWins'].replace(1,2)
scaled_data['blueWins'] = scaled_data['blueWins'].replace(0,1)

scaled_data.tail()
```

```python
scaled_data.to_csv('cleaned_league_data.txt',
                   float_format='%10.5f', sep=" ",
                   quoting=csv.QUOTE_NONE, escapechar=" ",
                   index=None, header=None)
```
